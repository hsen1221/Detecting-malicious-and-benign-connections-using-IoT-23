```python
import numpy as np
import pandas as pd
# Assign the filename
fpath = r"35.labeled"
# Import file using np.recfromcsv: d
narray = np.recfromcsv(fpath, encoding=None)

# Print out first n records of d 
print(narray)
# string to search in file
word = 'fields'
with open(fpath, 'r') as fp:
    # read all lines in a list
    lines = fp.readlines()
    for line in lines:
        # check if string present on a current line
        if line.find(word) != -1:
            print(word, 'string exists in file')
            print('Line Number:', lines.index(line))
            print('Line:', line)
            linefields = line
x = linefields.split("\t")
df = pd.concat([pd.Series(l.astype(str).split('\t')) for l in narray], axis=1).T
print(df)
```

```python
import pandas as pd
import re
import csv

# Assuming df is your DataFrame loaded in Jupyter

# 1. Clean column names
def clean_column_names(df):
    """Standardize column names for CSV export"""
    df.columns = [re.sub(r'\s+', '_', col.strip()).lower() for col in df.columns]
    df.columns = [re.sub(r'[^a-zA-Z0-9_]', '', col) for col in df.columns]
    return df

# Apply cleaning
df = clean_column_names(df)

# 2. Export with proper settings
df.to_csv('35.csv',
          index=False,
          encoding='utf-8-sig',
          quoting=csv.QUOTE_ALL,
          escapechar='\\')

print("CSV exported with clean column names")
```
